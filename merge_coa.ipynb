```python
headers = ["계정코드", "회사계정", "1차번역", "계정과목", "관리계정", "공시용계정", "합산계정", "분류", "구분"]
from_gongsi_headers = ["공시용계정", "합산계정", "분류", "구분"]
```

```python
# obj1: 회사계정 -> 관리계정
obj1_headers = ["계정코드", "1차번역", "관리계정"]

# obj2: 관리계정 -> 공시용계정
obj2_headers = ["계정코드", "관리계정", "공시용계정"]
```

```python
filenames = glob.glob(path + "\*.xlsx")
print("filenames:", filenames)
```

```python
dummyFinal = pd.DataFrame()
```

```python
# Comp: 계정과목의 상대적인 위치 (0 ~ 1)
for file in filenames:
    try:
        excel_file = pd.read_excel(file)
        excel_cols = excel_file.columns
        
        if('1차 번역' in excel_cols.values):
            excel_file.rename(columns={"1차 번역":"1차번역"}, inplace=True)
        if('계정 코드' in excel_cols.values):
            excel_file.rename(columns={"계정 코드":"계정코드"}, inplace=True)


        excel_file = excel_file.astype({'계정코드':'string'})
        df_sample = excel_file.loc[:, headers]
        df_sample['file_name'] = [file.replace("C://Users//jkwak020//Documents//CoA-Automapping/CoAs", "") for _ in range(len(df_sample))]
        df_sample['index'] = [i for i in range(len(df_sample))]
        df_sample['total_size'] = [len(df_sample) for _ in range(len(df_sample))]

        # display(df_sample)
        dummyFinal = dummyFinal.append(df_sample, ignore_index=True)
    except Exception as e:
        print(file, excel_cols)
```

```python
finalExcelSheet = dummyFinal
```

```python
print('Final Sheet: ')
display(finalExcelSheet)
```

```python
finalExcelSheet.dropna(inplace=True)
display(finalExcelSheet)
```

```python
finalExcelSheet.drop_duplicates(subset=headers,keep="first",inplace=True)
```
